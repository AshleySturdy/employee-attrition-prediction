### Missing Data
Check if there are missing values that need handling.

```python
# Check what columns have missing values
missing = df.isnull().sum()
missing[missing >0]
```

```python
# Calculate percentage of values missing
missing_df = pd.DataFrame({
    "Missing_Count": df.isnull().sum(),
    "Percentage": (df.isnull().sum() / len(df)) * 100})
missing_df[missing_df['Missing_Count'] > 0]
```

```python
# Number of columns with missing values
print(f"Number of columns with missing values: {len(missing_df[missing_df['Missing_Count'] > 0])}")
```

```python
# Find all rows with missing values
rows_with_missing = df[df.isnull().any(axis=1)]

print(f'Number of rows with missing values: {len(rows_with_missing)}')
```

```python
# Examine rows with missing values
rows_with_missing
```

```python
# Will drop these rows
df = df.dropna(how='any')
print(f'New shape: {df.shape}')
```

```python
# Verify no missing values
df.isnull().sum()
```

Identified 6 rows with missing values across almost all columns. These rows appeared to be data export artificacts and comprised less than 1% so they were removed from the dataset. The dataset now contains 1470 employees with no missing values.

### Data Types 

```python
# Check data types of variables
df.dtypes
```

```python
# Look at categorical variables
cat_cols = df.select_dtypes(include = ['object']).columns
print(f'Categorical columns: {list(cat_cols)}')
```

```python
# Unique values for categorical variables
for col in cat_cols:
    print(f'\n{col}: {df[col].nunique()} unique values')
    print(df[col].value_counts())
```

```python
# Convert Age to numeric
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
# Check for NaNs
df['Age'].isnull().sum()
```

```python
# Remove Over18 column - it is constant feature
print(df['Over18'].nunique())
df = df.drop('Over18', axis=1)
print(f'New shape: {df.shape}')
```

```python
# Get numeric columns
num_cols = df.select_dtypes(include = ['float64', 'int64']).columns
num_cols
```

```python
# Check for low cardinality - may intake variable should be categorical
for col in num_cols:
    n_unique = df[col].nunique()
    print(f'{col}: {n_unique} unique values')
    if n_unique <=10:
        print(f' Values: {sorted(df[col].unique())}')
    print()
```

```python
# Remove EmployeeCount and StandardHours - have constant value
df = df.drop(['EmployeeCount', 'StandardHours'], axis=1)
print(f'New shape: {df.shape}')
```

```python
# Remove EmployeeNumber as it is only an ID number without predictive value
df = df.drop('EmployeeNumber', axis=1)
```

```python
# Print categorical and numeric variables
cat_cols = df.select_dtypes(include = ['object']).columns
print(f'Number of categorical variables: {len(cat_cols)}')
print(f'List of categorical variables: {cat_cols}')

num_cols = [c for c in df if c not in cat_cols]
print(f'\nNumber of numeric variables: {len(num_cols)}')
print(f'List of numeric variables: {num_cols}')
```

**Data Type Adjustments**

EmployeeCount, StandardHours, and Over18: Removed due to constant value. EmployeeNumber removed as ID only and no predictive value.

Ordinal Variables: Several columns (Education, EnvironmentSatisfaction, JobInvolvement, JobLevel, JobSatisfaction, RelationshipSatisfaction, StockOptionLevel, WorkLifeBalance) contain ordinal categorical data encoded as integers. These represent ordered ratings/levels and will be kept as numeric for modeling, as the ordering contains meaningful information.

Age: Converted to numeric type.

### Target Variable
Attrition: Binary Indicator Variable

```python
# Evaluate target variable
df['Attrition'].value_counts()
```

```python
# Percentage for attrition
df['Attrition'].value_counts(normalize=True)*100
```

```python
# Convert 0/1 binary for modeling
df['Attrition'] = df['Attrition'].map({'Yes':1, 'No':0})
df['Attrition'].value_counts()
```

```python
# Change attrition data type to object
df['Attrition'] = df['Attrition'].astype('object')
```

**Target Variable Analysis**

The target variable, **Attrition**, shows class imbalance:
- **No**: 83.88% (1233 employees)
- **Yes**: 16.12% (237 employees)

There is a moderate class imbalance. Techniques to consider:
- SMOTE
- Class weights in model training
- Stratifiedt sampling in train-test split

Converted to binary indicator 0/1. 0 = No, 1 = Yes.

### Summary

Age was converted to numeric data type. EmployeeCount, StandardHours, and Over18 were removed for constant values. EmployeeNumber removed as ID only.

The target variable shows moderate class imbalance, which will be addressed during modeling.
